```python
import pandas as pd
import numpy as np
```

```python
# Read pkl file
pkl_path = 'academic_test.pkl'
pkl = pd.read_pickle(pkl_path)
df = pd.DataFrame(pkl)
df['improved_qn'] = np.nan
```

```python
def inspect_row(df, row_index):
    print('---------------------')
    print(f'Question {row_index}')
    print(f'Original Question: {df["text"].iloc[row_index]}')
    print(f'Query: {df["query"].iloc[row_index]}')


def insert_modified_question(df, row_index, question):
    
    df['improved_qn'].iloc[row_index] = question
    # print(f'Original Question: {df["text"].iloc[row_index]}')
    # print(f'Query: {df["query"].iloc[row_index]}')
    print(f'Improved Question: {df["improved_qn"].iloc[row_index]}')
    
```

## Instruction for re-writing question

- inspect_row(df, index): allows you to check the original question
- insert_modified_question(df, index, improved_question): allows you to insert your improved question into the dataframe
